```python
estimator = Pipeline([("imputer", Imputer(#missing_values="NaN",
                                          strategy="mean",
                                          axis=0)),
                      ("logistic",logReg(random_state = 0))])
                      #("forest", RandomForestRegressor(random_state=0,
                      #                                 n_estimators=100))])
```

```python
#LR1.fit(X,y)
estimator.fit(X_train,y_train)
```
